```python
import pyttsx3

def speak(text):
    engine = pyttsx3.init()   # 🔥 new engine every time (same as your code)
    engine.say(text)
    engine.runAndWait()
    engine.stop()


# WELCOME
print("Welcome to Smart Animal Classifier AI!")
speak("Welcome to Smart Animal Classifier AI")

print("Answer the questions with yes or no.")
speak("Answer the questions with yes or no")


# QUESTION 1
speak("Does the animal have 4 legs?")
legs = input("Does the animal have 4 legs? (yes/no): ").lower()
speak(legs + " entered")

# QUESTION 2
speak("Can it fly?")
fly = input("Can it fly? (yes/no): ").lower()
speak(fly + " entered")

# QUESTION 3
speak("Can it swim?")
swim = input("Can it swim? (yes/no): ").lower()
speak(swim + " entered")

# QUESTION 4
speak("Does it bark?")
bark = input("Does it bark? (yes/no): ").lower()
speak(bark + " entered")

# QUESTION 5
speak("Is it a pet animal?")
pet = input("Is it a pet animal? (yes/no): ").lower()
speak(pet + " entered")


# PREDICTION LOGIC
if legs == "yes" and bark == "yes" and pet == "yes":
    animal = "Dog"
elif legs == "yes" and pet == "yes":
    animal = "Cat"
elif swim == "yes" and legs == "no":
    animal = "Fish"
elif fly == "yes":
    animal = "Bird"
elif legs == "yes":
    animal = "Cow"
else:
    animal = "Snake"


# OUTPUT
print("\nAI Prediction:")
speak("AI Prediction")

print("The animal is likely a", animal)
speak("The animal is likely a " + animal)
```
